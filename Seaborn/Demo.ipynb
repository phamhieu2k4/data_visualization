Barplot function: 
A bar plot gives an estimate of the central tendency for a numeric variable with the height of each rectangle. It provides some indication of the uncertainty around that estimate using error bars. To build this plot, you usually choose a categorical column on the x-axis and a numerical column on the y-axis.

```python
res = sns.barplot(x = mtcars['cyl'], y = mtcars['mpg'])
plt.show()
```

```python
res = sns.barplot(x = 'cyl', y = 'mpg', data = mtcars)
plt.show()
```

```python
#check out some paletteL: https://seaborn.pydata.org/tutorial/color_palettes.html
res = sns.barplot(x = 'cyl', y = 'mpg', data = mtcars, palette = 'rocket')
plt.show()
```

Countplot function: The countplot() function in the Python Seaborn library returns the count of total values for each category using bars.

```python
sns.countplot(x = 'cyl', data = mtcars, palette = 'Set1')
```

```python
sns.countplot(y = 'gear', data = mtcars, palette = 'Set1')
```

```python
sns.countplot(x = 'gear', hue = 'cyl', data = mtcars, palette = 'Set1')
```

Distribution Plot: 
The Seaborn library supports the distplot() function that creates the distribution of any continuous data.

```python
sns.displot(mtcars.mpg, bins = 10, color = 'b')
```

```python
#kernel dendisty estimates
sns.displot(mtcars.mpg, color = 'g', kind = 'kde')
```

Heatmap: lets you visualize matrix-like data. The values of the variables are contained in a matrix and are represented as colors.

```python
numeric_mtcars = mtcars.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_mtcars.corr()

# Plot heatmap of correlation matrix
sns.heatmap(corr_matrix, cbar=True, linewidths=0.5)
```

Scatterplot function: helps you create plots that can draw relationships between two continuous variables.

```python
from sklearn import datasets

iris = datasets.load_iris()

sns.scatterplot(x = iris.data[:, 0], y = iris.data[:, 1], data = iris)
```

```python
sns.scatterplot(x = iris.data[:, 0], y = iris.data[:, 1], data = iris, hue = iris.target)
```

```python
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

iris_data['species'] = iris.target_names[iris.target]

# Plot linear regression for petal length and petal width
sns.lmplot(x='petal length (cm)', y='petal width (cm)', data=iris_data, hue='species', markers=["o", "s", "D"])
```
